```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#### Read the real time data from the source
df  =pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
df
```

```python
#### Filter the data by deleting the unnecessary row 36
df = df.iloc[:-1,1:]
df
```

```python
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active']
```

### Total positive cases based on the States of India

```python
max(df['positive'])
```

```python
from matplotlib.pyplot import figure
figure(num = None, figsize=(11,9), dpi=80, facecolor = 'w', edgecolor='k')
plt.barh(total_states, df['positive'], align = 'center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.4,0.8))
plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['positive'])+100)
plt.xlabel('Positive number of cases')
plt.title('Corona Virus Cases')
plt.show()
```

### Active Number of cases

```python
from matplotlib.pyplot import figure
figure(num = None, figsize=(11,9), dpi=80, facecolor = 'w', edgecolor='k')
plt.barh(total_states, df['new_active'], align = 'center', alpha=0.5, color=(1,0.5,0), edgecolor=(0.5,0.4,0.8))
plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['new_active'])+100)
plt.xlabel('New active number of cases')
plt.title('Corona Virus Cases')
plt.show()
```

### Death Cases 

```python
from matplotlib.pyplot import figure
figure(num = None, figsize=(11,9), dpi=80, facecolor = 'w', edgecolor='k')
plt.barh(total_states, df['death'], align = 'center', alpha=0.5, color=(0.5,0.5,1), edgecolor=(0.5,0.4,0.8))
plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['death'])+100)
plt.xlabel('Number of Death cases')
plt.title('Corona Virus Cases')
plt.show()
```
